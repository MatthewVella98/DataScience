# Web Scraper 

```python
import requests 
from bs4 import BeautifulSoup 

# Collect the github page
page = requests.get('https://github.com/trending')
print(page)

# Response 200 = Page was fetched succesfully 
```

```python
# Creating a BeautifulSoup object 

soup = BeautifulSoup(page.text, 'html.parser')

print(soup)      # This gives us the wkole HTML of the page fetched. 
```

## Extracting Data

#### Open the Github page. 
#### Clicking 'inspect anywhere', the body of our wanted data is the div tag <#div class="repo-list"> 

#### Each repository is deifned as a <#li>. We need to retrieve this. 

```python
# Using - The previously fetched: page AND The previously created: soup

# Get the repository list 
repo = soup.find(class_='repo-list')

# Find all the instances of the class (Should return 25 as shown in the page)
repo_list = repo.find_all(class_='col-9 text-gray my-1 pr-4')

print(len(repo_list))
```
